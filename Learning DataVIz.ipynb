Working with CSV 

```python
import csv 
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter

plt.style.use("fivethirtyeight")

with open('student-por.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    age_counter = Counter()
    for row in csv_reader:
        age_counter.update(row['school;sex;age;address;famsize;Pstatus;Medu;Fedu;Mjob;Fjob;reason;guardian;traveltime;studytime;failures;schoolsup;famsup;paid;activities;nursery;higher;internet;romantic;famrel;freetime;goout;Dalc;Walc;health;absences;G1;G2;G3'].split(';')[1])
#     row = next(csv_reader)
#     print(row['school;sex;age;address;famsize;Pstatus;Medu;Fedu;Mjob;Fjob;reason;guardian;traveltime;studytime;failures;schoolsup;famsup;paid;activities;nursery;higher;internet;romantic;famrel;freetime;goout;Dalc;Walc;health;absences;G1;G2;G3'].split(';')[1])


female_Count = age_counter['F']
male_Count = age_counter['M']
print('females: '+str(female_Count))
print('males: '+str(male_Count))

gender = ['Females','Males']
gender_data = [female_Count,male_Count]
# print(gender_data)
fig = plt.figure()
ax = fig.add_axes([0,0,1,2])
# ax.axis('equal')
plt.pie(gender_data,labels = gender,autopct='%1.2f%%')
plt.title('Gender Distribution of Males and Females')
plt.show()
```
